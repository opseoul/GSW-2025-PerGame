```python
import pandas as pd

df = pd.read_csv("GSW_2025_PerGame.csv")
df.head()
```

```python
# drop team total row
df = df[df["Player"] != "Team Totals"]

# convert numeric columns
for col in df. columns[3:-2]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# confirm cleaned
df.info()
df.describe().T
```

```python
# create custom metric columns
df["Scoring_Efficiency"] = df["PTS"] / df["FGA"]
df["All_Around_Efficiency"] = df["PTS"] + df["AST"] + df["TRB"] + df["STL"] + df["BLK"]

# Top 5 in each
top_efficiency = df[["Player", "Scoring_Efficiency"]].sort_values(by="Scoring_Efficiency", ascending=False).head(5)
top_all_around = df[["Player", "All_Around_Efficiency"]].sort_values(by="All_Around_Efficiency", ascending=False).head(5)

print("Top 5 Scoring Efficiency:")
print(top_efficiency)

print("\nTop 5 All-Around Efficiency:")
print(top_all_around)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Top Scorers
top_scorers = df.sort_values(by="PTS", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_scorers, x="PTS", y="Player", palette="viridis")
plt.title("Top 10 Scorers in GSW 2025 Season")
plt.xlabel("Points per Game")
plt.ylabel("")
plt.show()
```

```python
# Load and inspect
df = pd.read_csv("GSW_2025_PerGame.csv")

# Ensure 'Pos' is clean: drop rows where MP, PTS, or Pos are missing
df_clean = df[['MP', 'PTS', 'Pos']].dropna()

# Check for lingering NaNs or whitespace
df_clean['Pos'] = df_clean['Pos'].astype(str).str.strip()

# Confirm the unique values
print("Unique positions:", df_clean['Pos'].unique())
print("Any NaNs left?", df_clean['Pos'].isna().sum())  # Should be 0


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_clean, x="MP", y="PTS", hue="Pos")
plt.title("Minutes Played vs. Points Scored")
plt.xlabel("Minutes Per Game")
plt.ylabel("Points Per Game")
plt.legend(title="Position")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_clean.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - GSW Player Stats")
plt.show()
```

```python
# compare starters vs bench
df['Role'] = df['GS'].apply(lambda x: 'Starter' if x > 30 else 'Bench')
sns.boxplot(data=df, x='Role', y='PTS')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select performance features for clustering
features = ['PTS', 'MP', 'FG%', 'AST', 'STL', 'TRB']
df_cluster = df[features].copy()

# Drop rows with any NaNs in the features
df_cluster_clean = df_cluster.dropna().reset_index(drop=True)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster_clean)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster_clean['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualize the clusters (e.g., MP vs. PTS)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cluster_clean, x='MP', y='PTS', hue='Cluster', palette='Set1', s=100)
plt.title('Player Clusters based on Minutes and Points')
plt.xlabel('Minutes Per Game')
plt.ylabel('Points Per Game')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# Select relevant features
features = ['MP', 'PTS']
df_cluster = df[features].copy()

# Drop rows with NaNs in selected features
df_cluster = df_cluster.dropna()

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(X_scaled)

df_cluster.groupby('Cluster')[['MP', 'PTS']].mean()

```

```python
# using more performance features
features = ['MP', 'PTS', 'AST', 'TRB', 'FG%', 'STL', 'BLK', 'TOV']

# clean and scale the dataset
df_cluster = df[features].copy()
df_cluster = df_cluster.dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)

# kmeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(X_scaled)

# label clusters based on stats
df_cluster.groupby('Cluster')[features].mean()
# Map numeric cluster IDs to labels (you might need to adjust these based on the actual means)
label_map = {
    0: 'High-Impact Starter',
    1: 'Role Player',
    2: 'Bench/Development'
}
df_cluster['Role'] = df_cluster['Cluster'].map(label_map)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df_cluster, x='MP', y='PTS', hue='Role', palette='Set2')
plt.title('GSW Player Roles Based on Performance Clustering')
plt.xlabel('Minutes Per Game')
plt.ylabel('Points Per Game')
plt.legend(title='Player Role')
plt.show()
```
